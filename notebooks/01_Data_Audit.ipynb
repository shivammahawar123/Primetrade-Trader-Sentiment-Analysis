# Data Audit

This notebook performs an initial audit of the raw datasets to assess their structure, quality, and readiness for analysis.

## Load Libraries and Datasets

Import the required library and load the raw datasets.

```python
import pandas as pd

# Display all columns
pd.set_option("display.max_columns", None)

# Load datasets
historical_data = pd.read_csv("../data/raw/historical_data.csv")
fear_greed_index = pd.read_csv("../data/raw/fear_greed_index.csv")
```

## Dataset Preview

Review the first few records of each dataset to understand their structure and verify that the data has been loaded correctly.

```python
print("Historical Data")
display(historical_data.head())

print("\nFear & Greed Index")
display(fear_greed_index.head())
```

## Dataset Information

Review the dimensions, column names, data types, and memory usage of each dataset.

```python
print("Historical Data")
historical_data.info()

print("\n" + "=" * 60 + "\n")

print("Fear & Greed Index")
fear_greed_index.info()
```

## Dataset Dimensions

Review the number of rows and columns in each dataset.

```python
print("Historical Data Shape:", historical_data.shape)
print("Fear & Greed Index Shape:", fear_greed_index.shape)
```

## Column Names

Review the column names of each dataset.

```python
print("Historical Data Columns")
print(historical_data.columns.tolist())

print("\nFear & Greed Index Columns")
print(fear_greed_index.columns.tolist())
```

## Missing Values

Check for missing values across all columns in both datasets.

```python
print("Historical Data")
display(historical_data.isna().sum().to_frame("Missing Values"))

print("\nFear & Greed Index")
display(fear_greed_index.isna().sum().to_frame("Missing Values"))
```

## Duplicate Records

Check for duplicate rows in each dataset.

```python
print("Historical Data Duplicates:", historical_data.duplicated().sum())
print("Fear & Greed Index Duplicates:", fear_greed_index.duplicated().sum())
```

## Data Types

Review the data types of all columns to identify any inconsistencies requiring validation.

```python
display(historical_data.dtypes.to_frame("Data Type"))

display(fear_greed_index.dtypes.to_frame("Data Type"))
```

## Date and Time Columns

Review the date and time columns before validating their formats and consistency.

```python
print("Historical Data")
display(historical_data[["Timestamp IST", "Timestamp"]].head())

print("\nFear & Greed Index")
display(fear_greed_index[["date", "timestamp"]].head())
```

## Date Range

Review the date range covered by each dataset.

```python
print("Historical Data")
print("Timestamp IST")
print("Earliest:", historical_data["Timestamp IST"].min())
print("Latest  :", historical_data["Timestamp IST"].max())

print("\nFear & Greed Index")
print("Date")
print("Earliest:", fear_greed_index["date"].min())
print("Latest  :", fear_greed_index["date"].max())
```

## Unique Values

Review the unique values in categorical columns to identify inconsistencies and invalid entries.

```python
categorical_columns = [
    "Coin",
    "Side",
    "Direction",
    "Crossed"
]

for column in categorical_columns:
    print(f"\n{column} ({historical_data[column].nunique()} unique values)")
    print(sorted(historical_data[column].unique()))

print("\nClassification")
print(sorted(fear_greed_index["classification"].unique()))
```

## Numeric Value Validation

Review the numeric columns to identify invalid values and verify business consistency.

```python
numeric_columns = [
    "Execution Price",
    "Size Tokens",
    "Size USD",
    "Start Position",
    "Closed PnL",
    "Fee"
]

historical_data[numeric_columns].agg(["min", "max"]).T
```

## Zero Value Validation

Review numeric columns for zero values that may require validation.

```python
numeric_columns = [
    "Execution Price",
    "Size Tokens",
    "Size USD",
    "Start Position",
    "Closed PnL",
    "Fee"
]

zero_counts = (historical_data[numeric_columns] == 0).sum().to_frame("Zero Values")
display(zero_counts)
```

## Timestamp Consistency

Review the uniqueness of timestamp columns to identify potential inconsistencies.

```python
print("Unique values in Timestamp IST:", historical_data["Timestamp IST"].nunique())
print("Unique values in Timestamp    :", historical_data["Timestamp"].nunique())

display(
    historical_data["Timestamp"]
    .value_counts()
    .sort_index()
    .rename("Record Count")
    .to_frame()
)
```

## Timestamp Coverage

Review the date range covered by each timestamp column.

```python
print("Timestamp IST")
print("Earliest:", historical_data["Timestamp IST"].min())
print("Latest  :", historical_data["Timestamp IST"].max())

print("\nTimestamp")
print("Earliest:", historical_data["Timestamp"].min())
print("Latest  :", historical_data["Timestamp"].max())

print("\nFear & Greed Date")
print("Earliest:", fear_greed_index["date"].min())
print("Latest  :", fear_greed_index["date"].max())
```

## Trade Direction Consistency

Review the relationship between trade side and direction to identify inconsistent or unexpected combinations.

```python
direction_consistency = (
    historical_data
    .groupby(["Side", "Direction"])
    .size()
    .reset_index(name="Trade Count")
    .sort_values(by=["Side", "Trade Count"], ascending=[True, False])
)

display(direction_consistency)
```

## Data Audit Findings

Summarize the key observations identified during the data audit.

- Both datasets were successfully loaded and inspected.
- No missing values or duplicate records were identified in either dataset.
- The `Timestamp` column contains only 7 unique values, whereas `Timestamp IST` contains 27,977 unique values. This inconsistency should be investigated before using the `Timestamp` column for time-based analysis.
- The `Coin` column contains 77 asset identifiers prefixed with `@`. Their business meaning should be verified before analysis.
- The `Direction` column includes multiple trade event types such as position openings, position closures, liquidations, settlements, and spot dust conversions, indicating that it represents trade actions beyond simple buy and sell transactions.
- Zero-value trade records were associated with **Spot Dust Conversion** events and appear to represent valid exchange activity rather than missing or invalid data.
- Negative fee values were observed and may represent exchange rebate mechanisms rather than data quality issues.
