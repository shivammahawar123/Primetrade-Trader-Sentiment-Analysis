# Data Cleaning

Prepare the raw datasets for analysis by addressing the data quality issues identified during the data audit. The cleaned datasets will be used in subsequent project stages.

## Import Libraries

Import the required libraries for data cleaning and preprocessing.

```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
pd.set_option("display.float_format", "{:.2f}".format)
```

## Load Datasets

Load the raw datasets into DataFrames for cleaning.

```python
historical_df = pd.read_csv("../data/raw/historical_data.csv")
sentiment_df = pd.read_csv("../data/raw/fear_greed_index.csv")
```

## Create Working Copies

Create copies of the raw datasets to preserve the original data throughout the cleaning process.

```python
historical_clean = historical_df.copy()
sentiment_clean = sentiment_df.copy()
```

## Data Types

Review the data types identified during the data audit before applying the required corrections.

```python
display(historical_clean.dtypes.to_frame(name="Data Type"))
```

```python

display(sentiment_clean.dtypes.to_frame(name="Data Type"))
```

## Data Type Conversion

Convert date and time fields to appropriate datetime formats for consistent processing and analysis.

```python
historical_clean["Timestamp IST"] = pd.to_datetime(
    historical_clean["Timestamp IST"],
    format="%d-%m-%Y %H:%M",
    errors="coerce"
)

historical_clean["Timestamp"] = pd.to_datetime(
    historical_clean["Timestamp"],
    unit="ms",
    errors="coerce"
)

sentiment_clean["date"] = pd.to_datetime(
    sentiment_clean["date"],
    errors="coerce"
)

sentiment_clean["timestamp"] = pd.to_datetime(
    sentiment_clean["timestamp"],
    unit="s",
    errors="coerce"
)
```

### Historical Data

```python
display(historical_clean.dtypes.to_frame(name="Data Type"))
```

### Fear & Greed Data

```python
display(sentiment_clean.dtypes.to_frame(name="Data Type"))
```

## Missing Values

Review the missing values identified during the data audit before applying the required treatments.

### Historical Data

```python
historical_missing = (
    historical_clean.isna()
    .sum()
    .rename("Missing Values")
    .to_frame()
    .query("`Missing Values` > 0")
)

display(historical_missing)
```

### Fear & Greed Data

```python
sentiment_missing = (
    sentiment_clean.isna()
    .sum()
    .rename("Missing Values")
    .to_frame()
    .query("`Missing Values` > 0")
)

display(sentiment_missing)
```

## Duplicate Records

Verify that both datasets do not contain duplicate records before proceeding with further cleaning.

```python
duplicate_summary = pd.DataFrame({
    "Dataset": ["Historical Data", "Fear & Greed Data"],
    "Duplicate Records": [
        historical_clean.duplicated().sum(),
        sentiment_clean.duplicated().sum()
    ]
})

display(duplicate_summary)
```

## Data Consistency

Review the categorical fields to ensure values are standardized and consistent across the datasets.

### Historical Data

```python
historical_columns = ["Coin", "Side", "Direction", "Crossed"]

for column in historical_columns:
    print(f"\n{column}")
    
    display(
        pd.DataFrame(
            sorted(historical_clean[column].dropna().unique()),
            columns=[column]
        )
    )
```

### Fear & Greed Data

```python
display(
    pd.DataFrame(
        sorted(sentiment_clean["classification"].dropna().unique()),
        columns=["Classification"]
    )
)
```

## Export Cleaned Data

Save the cleaned datasets for use in the subsequent project notebooks.

```python
historical_clean.to_csv(
    "../data/processed/historical_data_clean.csv",
    index=False
)

sentiment_clean.to_csv(
    "../data/processed/fear_greed_index_clean.csv",
    index=False
)

print("Cleaned datasets saved successfully.")
```

The datasets were successfully prepared for analysis by applying the required data cleaning procedures.

- Verified the absence of missing values and duplicate records.
